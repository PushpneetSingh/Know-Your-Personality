#Actual 5 cluster

```
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
df = sd.iloc[:,:50]
ssd = []
K = range(2,15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(df)
    ssd.append(km.inertia_)
# Plot sum of squared distances / elbow method
plt.figure(figsize=(10,6))
plt.plot(K, ssd, 'bx-')
plt.xlabel('k')
plt.ylabel('ssd')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```
# Create and fit model
from sklearn.cluster import KMeans
df = sd.iloc[:,:50]
kmeans = KMeans(n_clusters=5)
model = kmeans.fit(df)
pred = model.labels_
sd["cluster"] = pred
sd.head()
```

```
df["cluster"] = pred
df.head()
df.groupby('cluster').mean()
```

```
from sklearn.decomposition import PCA
 
pca = PCA(n_components=2)
pca_fit = pca.fit_transform(df)
 
df_pca = pd.DataFrame(data=pca_fit, columns=['PCA1', 'PCA2'])
df_pca['Clusters'] = pred
df_pca.head()
```

```
plt.figure(figsize=(10,10))
sns.scatterplot(data=df_pca, x='PCA1', y='PCA2', hue='Clusters', palette='Set2', alpha=0.8)
plt.title('Personality Clusters after PCA');
```

```
# Summing up the different questions groups via factor analysis
col_list2 = list(dataframe)
ext = col_list2[0:10]
est = col_list2[10:20]
agr = col_list2[20:30]
csn = col_list2[30:40]
opn = col_list2[40:50]

dataframeFA = pd.DataFrame()
dataframeFA['extroversionFA'] = dataframe[ext].sum(axis=1)/10
dataframeFA['neuroticFA'] = dataframe[est].sum(axis=1)/10
dataframeFA['agreeableFA'] = dataframe[agr].sum(axis=1)/10
dataframeFA['conscientiousFA'] = dataframe[csn].sum(axis=1)/10
dataframeFA['openFA'] = dataframe[opn].sum(axis=1)/10

dataframeFA
```

```
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
df2 = dataframeFA.iloc[:,:]
ssd2 = []
K = range(2,15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(df2)
    ssd2.append(km.inertia_)
# Plot sum of squared distances / elbow method
plt.figure(figsize=(10,6))
plt.plot(K, ssd2, 'bx-')
plt.xlabel('k')
plt.ylabel('ssd')
plt.title('Elbow Method For Optimal k after factor analysis')
plt.show()
```

```
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
# Create and fit model
kmeans = KMeans(n_clusters=5)
model2 = kmeans.fit(dataframeFA)
pred2 = model2.labels_
dataframeFA["cluster"] = pred2

dataframeFA
```

```
 result1  = dataframeFA.groupby('cluster').mean() #after FA 5 cluster
```

```
pca = PCA(n_components=2)
pca_fit = pca.fit_transform(dataframeFA)
 
df_pca = pd.DataFrame(data=pca_fit, columns=['PCA1', 'PCA2'])
df_pca['Clusters'] = pred2
df_pca.head()

plt.figure(figsize=(10,10))
sns.scatterplot(data=df_pca, x='PCA1', y='PCA2', hue='Clusters', palette='Set2', alpha=0.8)
plt.title('Personality Clusters after PCA');
```
